```python
# Counting the total number of words 
hme['word_count'] = hme['reviews'].apply(lambda x: len(str(x).split(" ")))
hme[['reviews','word_count']].head()
```

#  <span style='background : cyan'>LDA ( Latent Dirichlet Allocation) </span>

### 1.LDA is a generative statistical model that explains a set of observations through unobserved groups
### 2.It builds a topic per document model and words per topic model, modeled as Dirichlet distributions. 
### 3.Each document is modeled as a multinomial distribution of topics and each topic is modeled as a          multinomial distribution of words.

<img src="https://editor.analyticsvidhya.com/uploads/4519707623a1e07ae153046bc387e0136a65f.image001-min.png" /> 

```python
# Creating Corpus to train LDA model using lemmatized word
doc=hme['tok_word']
doc
```

```python
# Creating dictionary 
dictionary=corpora.Dictionary(doc)
```

```python
# Converting the corpus to Document_term_matrix using dictionary
document_term_matrix=[dictionary.doc2bow(word) for word in doc] 
```

```python
# Creating model using gensim library
Lda=gensim.models.ldamodel.LdaModel
```

```python
# Training the LDA model on the Document_term_matrix
No_of_topics=int(input("Enter a number"))
ldamodel=Lda(corpus=document_term_matrix,num_topics=No_of_topics,id2word=dictionary)
```

```python
for idx, topic in ldamodel.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(idx, topic))
```
